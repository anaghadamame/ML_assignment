# Train Dataset:

```python
df2=pd.read_csv("train.csv")
df2.head(2)
```

```python
df2.head()
```

```python
#checking non numberical values from the data set
df_obj=pd.DataFrame()
cnt=1
for col in df2.columns:
    if(df2[col].dtype=='O'):
          df_obj[col]=df2[col]
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
df_obj=pd.DataFrame(imp.fit_transform(df_obj),columns=df_obj.columns,index=df_obj.index)
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
#df_obj=df_obj.columns.to_numpy().tolist()
for col in df_obj.columns[0:]:
    df_obj[col]= lb_make.fit_transform(df_obj[col])
df_obj.head(3)
```

```python
df_num=pd.DataFrame()
cnt=1
for col in df2.columns:
    if(df2[col].dtype=='int64' or df2[col].dtype=='float64'):
          df_num[col]=df2[col]
#checking  numberical values from the data set
cnt=1
for col in df_num.columns:
    #if(df1[col].dtype=='int64' or df1[col].dtype=='float64'):
        if(df_num[col].isnull().sum()!=0):
            df_num[col]=df_num[col].replace(np.NaN,df_num[col].mean())
df_num.head(3)
```

```python
#Combining numerical & catagorical dataset
df2=pd.concat([df_obj,df_num],axis=1)
df2.head(2)
```

# Feature Selection:

```python
X = df2.iloc[:,:]  #independent columns
y = df2.iloc[:,-1]    #target column i.e price range
```

```python
#apply SelectKBest class to extract top 40 best features
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
kbest_model = SelectKBest(score_func =chi2, k=30)
kbest_model.fit(X,y)
```

```python
dict1= {
    'Columns':X.columns,
    'Scores':kbest_model.scores_
}

feature_scores_df = pd.DataFrame(dict1)
```

```python
feature_scores_df
```

```python
df_f=feature_scores_df.nlargest(40,'Scores')
#df_f.columns
c=df_f['Columns'].values
```

```python
df2_c=df2[:].copy()
df1_cop=df1[:].copy()
df1_cop.head()
```

```python
clm=list()
df2_c1=pd.DataFrame()
for col in df2_c.columns[:]:
    for i in c:
        if col ==i:
            clm.append(col)
            df2_c1[col]=df2_c[col]
print(len(clm))
df2_c1.head()
```

```python
clm=list()
df1_c1=pd.DataFrame()
for col in df1_cop.columns[:]:
    for i in c:
        if col ==i:
            clm.append(col)
            df1_c1[col]=df1_cop[col]
print(len(clm))
df1_c1.head()
```

```python
df2.head(2)
```

```python
x_test=df1_c1.iloc[:,:].values #Test data (After feature selection)
x_train=df2_c1.iloc[:,1:].values #train data (After feature selection)
y_train=df2.iloc[:,-1].values #SalePrice dependent variable
```

# Cross Validation:

```python
from sklearn.model_selection import KFold,cross_val_score
from sklearn.metrics import make_scorer, r2_score,accuracy_score
#from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
def test_model(model,x_train=x_train,y_train=y_train):
    cv=KFold(n_splits=3,shuffle=True,random_state=0)
    r2=make_scorer(r2_score)
    r2_val_score=cross_val_score(model,x_train,y_train,cv=cv,scoring=r2)
    score=[r2_val_score.mean()]
    return score
```

```python
def test_model_acc(model,x_train=x_train,y_train=y_train):
    cv=KFold(n_splits=3,shuffle=True,random_state=0)
    r2=make_scorer(accuracy_score)
    r2_val_score=cross_val_score(model,x_train,y_train,cv=cv,scoring=r2)
    score=[r2_val_score.mean()]
    return score
```

```python
import sklearn.linear_model as linear_model
LR = linear_model.LinearRegression()
print(test_model(LR))
```

```python
rgd=linear_model.Ridge()
test_model(rgd)
```

```python
lasso=linear_model.Lasso()
test_model(lasso)
#test_model_acc(lasso)
```

```python
#XGBoost
import xgboost
xgb_reg = xgboost.XGBRegressor(bbooster='gbtree',random_state=0)
test_model(xgb_reg)
```

```python
xgb_reg.fit(x_train,y_train)
y_pre=xgb_reg.predict(x_test)
print(len(y_pre))
print(y_pre)
```

```python
output = pd.DataFrame({'HouseId': HouesID, 'SalePrice': y_pre})
output.to_csv('Output.csv', index=False)
```
